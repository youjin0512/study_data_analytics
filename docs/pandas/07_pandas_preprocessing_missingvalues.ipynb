## 전처리(Preprocessing)
- 정형 데이터 기준 : 이상치/결측치에 대한 값 처리 방법 연구

### 결측치(missin values) : Null, NaN(None a Number : a가 없다), None, "" <값이 없다>
- 처리 방식
- 1) 결측치 확인 : pd.isnull().sum()
- 2) 결측치 처리
  -  결측치 삭제(제거) : pd.dropna(subset=[ ], 행 단위 적용)
  -  결측치 치환 : 대표값 or 머신러닝 예측값으로 치환, 열 단위 적용



```python
import pandas as pd
import numpy as np
```

```python
# 데이터셋 생성
data = {
    "수치형": [1, None, 3, np.nan, 5],
    "범주형": [None, 'B', 'C', np.nan, 'E'],
    "datetime": pd.date_range(start='2021-01-01', periods=4).insert(3, np.nan)
}
df_temp = pd.DataFrame(data)
df_temp
```

### 1) 결측치 확인

```python
df_temp.info()
```

```python
# df_temp.isnull()    # Null이 들어간 부분은 'True'로 출력됨
df_temp.isnull().sum()    # 각 컬럼별로 결측치 총합 표시
```

### 2) 결측치 처리

#####  (1) 결측치 제거 : '행' 단위 적용




```python
# df_temp.dropna()   # na =~ NaN       ** =~ : (~와유사하다)
df_temp.dropna(inplace=True)   # na =~ NaN       ** =~ : (~와유사하다)
df_temp
```

```python
df_temp.dropna(subset=['범주형'])  # 특정 컬럼 지정할 때 사용
```

##### (2) 결측치 치환(채우기)
 - 1) 대표값을 설정해 채움
 - 2) 머신러닝 (KNN) 기술 사용해 채움

```python
df_temp.describe()
# df_temp.describe().loc['mean', '수치형']
```

```python
df_temp['수치형'].fillna(4.0, inplace=True)
```

```python
df_temp
```

```python
df_temp.describe(include=object)   # 범주형 가져오기
```

```python
df_temp['범주형'].value_counts()
```

```python
df_temp['범주형'].fillna('E,', inplace=True)
df_temp
```

```python
df_temp.describe(include='datetime')  # 분포에 영향 많으므로 되도록 삭제
```

```python
df_temp.dropna(inplace=True)
df_temp
```
