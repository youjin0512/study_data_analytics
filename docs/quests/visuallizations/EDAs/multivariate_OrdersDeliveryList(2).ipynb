# 2. 모바일웹을 통해 주문된 상품 중, 상품구매금액이 평균 이상인 주문들을 찾으세요.

```python
# 전체 '상품구매금액'의 평균 구하기 mean
average_purchase_amount = df_LCOCT['상품구매금액'].mean() # 평균금액 69368.14445228774
# '주문경로'가 '모바일웹'이고 '상품구매금액'이 평균 이상인 주문
df_filtered = df_LCOCT[(df_LCOCT['주문경로'] == '모바일웹') & (df_LCOCT['상품구매금액'] >= average_purchase_amount)]
df_filtered
```

# 3. 2019년 1월 7일 이전에 배송이 시작된 주문들을 찾으세요.

```python
# '오전'과 '오후'를 'AM'과 'PM'으로 변환합니다.
df_LCOCT['배송시작일'] = df_LCOCT['배송시작일'].str.replace('오전', 'AM')
df_LCOCT['배송시작일'] = df_LCOCT['배송시작일'].str.replace('오후', 'PM')

# '배송시작일'을 날짜 형태로 변환
df_LCOCT['배송시작일'] = pd.to_datetime(df_LCOCT['배송시작일'], format='%Y-%m-%d %p %I:%M', errors='coerce')
# %p : am or pm
# %I : 12시간의 형식
# %M : 분

# 2019년 1월 7일 이전에 배송이 시작된 주문 찾기
df_LCOCT_filtered = df_LCOCT[df_LCOCT['배송시작일'] < '2019-01-07']

df_LCOCT_filtered
```

# 4. 배송 시작일과 배송 완료일이 3일 이상 차이 나는 주문들을 찾으세요.

```python
# '배송시작일'과 '배송완료일'을 날짜 형태로 변환
df_LCOCT['배송시작일'] = pd.to_datetime(df_LCOCT['배송시작일'], format='%Y-%m-%d %p %I:%M', errors='coerce')
df_LCOCT['배송완료일'] = pd.to_datetime(df_LCOCT['배송완료일'], format='%Y-%m-%d %p %I:%M', errors='coerce')

# 배송 시간이 3일 이상 걸린 주문을 찾기
df_LCOCT_filtered = df_LCOCT[(df_LCOCT['배송완료일'] - df_LCOCT['배송시작일']).dt.days >= 3]

df_LCOCT_filtered
```
